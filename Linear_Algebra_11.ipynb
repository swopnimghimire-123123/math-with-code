#  Abstract Vector Spaces

###  Idea
- A **vector space** is a set of objects called **vectors**, along with two operations:
  1. **Vector addition**: u + v  
  2. **Scalar multiplication**: c * v  

- These operations satisfy certain rules (axioms), such as:
  - Commutativity: u + v = v + u  
  - Associativity: (u + v) + w = u + (v + w)  
  - Existence of zero vector: v + 0 = v  
  - Existence of additive inverse: v + (-v) = 0  
  - Compatibility with scalars: a(bv) = (ab)v  
  - Distributivity: a(u + v) = au + av  

---

###  Examples of Vector Spaces
1. **Standard 2D and 3D vectors**: ℝ², ℝ³  
2. **Polynomials of degree ≤ n**  
3. **Functions**: f: ℝ → ℝ  
4. **Matrices**: All m×n matrices form a vector space  
5. **Sequences**: Infinite sequences of numbers  

---

###  Key Concepts
- **Subspace**: A subset of a vector space that is itself a vector space.  
- **Basis**: A minimal set of vectors that spans the space.  
- **Dimension**: Number of vectors in a basis.  
- **Linear combination**: Sum of scalars times vectors in the space.  

 **Intuition**:  
Vector spaces generalize the idea of “vectors” beyond just arrows in space—they capture **any objects that can be added and scaled** while following the same rules.



```python
import numpy as np
# step1: vector addition
v1 = np.array([1,2,4])
v2 = np.array([5,6,7])

sum = v1 + v2
print("v1 + v2:",sum)

# step2: scaler multiplication
scalar = 3
scaled_vec = scalar * v1

print("Scaled vector:",scaled_vec)

# step3: Linear Combination
a, b = 2, -1
lin_comb = a*v1 + b*v2

print("Linear Combination:",lin_comb)

# step4: polynomial vector space (degree < 2)
p1 = np.array([2, 3, -1])
p2 = np.array([1, -1, 4])
poly_sum = p1 + p2
scaled_poly = 2 * p1

print("p1 + p2 =", poly_sum)
print("2 * p1 =", scaled_poly,"\n")

# Step 5: Matrix vector space
M1 = np.array([[1, 2], [3, 4]])
M2 = np.array([[0, 1], [1, 0]])
matrix_sum = M1 + M2
scaled_matrix = -1 * M1

print("M1 + M2 =\n", matrix_sum,"\n")
print("-1 * M1 =\n", scaled_matrix,"\n")

# Step 6: Checking subspace
# A subspace must be closed under addition and scalar multiplication
subspace_check = (v1 + v2, 3*v1)

print("Sum:", subspace_check[0])
print("Scalar multiple:", subspace_check[1])
```

### Deep Intuition

- Vector addition = combining
directions.

- Scalar multiplication = stretching/shrinking along the same direction.

- Linear combination = generating all possible vectors in the space.

- Polynomials, matrices, functions = all act like vectors as long as they obey addition and scaling rules.

- Subspace = any subset closed under these operations, forming a smaller vector space.

#  Functions as Vectors

###  Idea
- Functions can act like vectors if we define **addition** and **scalar multiplication** properly.
- Example vector space: all continuous functions f: ℝ → ℝ.

###  Operations
1. **Function addition**: (f + g)(x) = f(x) + g(x)  
2. **Scalar multiplication**: (c * f)(x) = c * f(x)  

- These operations satisfy vector space axioms (commutativity, associativity, zero vector, etc.).

###  Intuition
- Think of each function as an “arrow” in an **infinite-dimensional space**.  
- Adding functions = combining their effects at every point x.  
- Scaling functions = stretching/shrinking their output.


```python
#  Functions as Vectors in Python
import numpy as np
import matplotlib.pyplot as plt

# Define functions as Python functions
def f(x):
    return x**2

def g(x):
    return np.sin(x)

# Step 1: Function addition
def sum_func(x):
    return f(x) + g(x)

# Step 2: Scalar multiplication
def scaled_func(x):
    return 2 * f(x)

# Step 3: Evaluate at points
x = np.linspace(-2, 2, 100)
plt.plot(x, f(x), label='f(x) = x^2')
plt.plot(x, g(x), label='g(x) = sin(x)')
plt.plot(x, sum_func(x), label='f(x)+g(x)')
plt.plot(x, scaled_func(x), label='2*f(x)')
plt.legend()
plt.title("Functions as Vectors")
plt.show()
```
