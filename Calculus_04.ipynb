# Trig Derrivative
### The Meaning of Trigonometric Derivatives

1. A derivative tells us **how fast a function is changing** at any point.

2. For trigonometric functions like sin(x), cos(x), and tan(x), this means:
   - How quickly the angle change (x) affects the function’s output.

3. On the unit circle:
   - sin(x) is the vertical (y) coordinate.
   - cos(x) is the horizontal (x) coordinate.
   - tan(x) is the slope of the radius (y/x).

4. The derivative of a trig function tells us the **instantaneous rate of change** of these coordinates as the angle sweeps around the circle.

   - d/dx[sin(x)] means: “How fast is the y-coordinate changing as the angle increases?”
   - d/dx[cos(x)] means: “How fast is the x-coordinate changing as the angle increases?”
   - d/dx[tan(x)] means: “How fast is the slope y/x changing as the angle increases?”

 In short: Trig derivatives describe **the speed of change of circular motion** projected onto x (cos), y (sin), or slope (tan).


### Geometric Intuition of Trigonometric Derivatives

1. **f(x) = sin(x):**
   - Imagine a point moving counterclockwise around the unit circle.
   - At angle x, the y-coordinate = sin(x).
   - The slope of sin(x) at that point is the x-coordinate of the unit circle, which is cos(x).
   -  Therefore, d/dx [sin(x)] = cos(x).

2. **f(x) = cos(x):**
   - On the unit circle, the x-coordinate = cos(x).
   - Its slope comes from the negative y-coordinate, which is -sin(x).
   -  Therefore, d/dx [cos(x)] = -sin(x).

3. **f(x) = tan(x):**
   - tan(x) = sin(x)/cos(x).
   - Using the quotient rule, the derivative becomes sec^2(x).
   - Geometrically, this reflects how tan(x) grows steeper as cos(x) → 0.
   -  Therefore, d/dx [tan(x)] = sec^2(x).


```python
import sympy as sp
# SymPy (Symbolic Python) is a Python library for symbolic mathematics.

x = sp.symbols('x')

# Define trig functions
f_sin = sp.sin(x)
f_cos = sp.cos(x)
f_tan = sp.tan(x)
f_csc = sp.csc(x)
f_sec = sp.sec(x)
f_cot = sp.cot(x)

# Derivatives
d_sin = sp.diff(f_sin, x)
d_cos = sp.diff(f_cos, x)
d_tan = sp.diff(f_tan, x)
d_csc = sp.diff(f_csc, x)
d_sec = sp.diff(f_sec, x)
d_cot = sp.diff(f_cot, x)

# Printing results
print("d/dx sin(x) =", d_sin)
print("d/dx cos(x) =", d_cos)
print("d/dx tan(x) =", d_tan)
print("d/dx csc(x) =", d_csc)
print("d/dx sec(x) =", d_sec)
print("d/dx cot(x) =", d_cot)
```

### Derivative Rules

1. **Sum Rule**: d/dx [f(x) + g(x)] = f'(x) + g'(x)  
2. **Product Rule**: d/dx [f(x) * g(x)] = f'(x)*g(x) + f(x)*g'(x)  
3. **Chain Rule**: d/dx [f(g(x))] = f'(g(x)) * g'(x)

**Intuition**:
- Sum Rule → derivative distributes over addition.  
- Product Rule → derivative of product = “first * derivative of second + second * derivative of first.”  
- Chain Rule → derivative of composition = derivative of outer evaluated at inner * derivative of inner.


```python
import sympy as sp

# define symbol
x = sp.symbols("x")

# example functions
f = x**2    # f(x)
g = sp.sin(x)   # g(x)

# sum rule
sum_rule = sp.diff(f+g,x)

# product rule
product_rule = sp.diff(f * g ,x)

# chain rule
chain_rule = sp.diff(sp.sin(x**2),x)

# Printing the derivatives
print("Sum Rule: d/dx [x^2 + sin(x)] =", sum_rule)
print("Product Rule: d/dx [x^2 * sin(x)] =", product_rule)
print("Chain Rule: d/dx [(sin(x))^2] =", chain_rule)
```
