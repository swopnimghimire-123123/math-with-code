## 1. Random Variables

` Definition:`

A random variable assigns a numeric value to the outcome of a random experiment.

Discrete: takes countable values (e.g., number of heads in coin flips).

Continuous: takes infinite values in a range (e.g., height, weight).

### 2. Probability Distributions

PMF (Discrete): probability of each outcome.

PDF (Continuous): probability density over a range.

$\sum P(X=x) = 1$ (Discrete)
$\int f(x) dx = 1$ (Continuous)


### 3. Binomial Distribution (Discrete)

$P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}$

$n$ = number of trials

$p$ = probability of success

$k$ = number of successes

Example: Toss a coin 10 times, probability of 6 heads.

```python
from scipy.stats import binom

n,p,k = 10, 0.5, 6
prob_6_heads = binom.pmf(k,n,p)
print(prob_6_heads)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

plt.bar(x, pmf)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.show()
```

### 4. Normal Distribution (Continuous)

 Formula:

$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

$\mu$ = mean (center)

$\sigma$ = standard deviation (spread)

Example: Heights of people often follow a normal distribution.

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

mu, sigma = 0, 1
x = np.linspace(-4, 4, 100)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf)
plt.title("Normal Distribution (μ=0, σ=1)")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid()
plt.show()
```

```python
###  Simulation Example
### stimulate 1000 coin flips, compare empirical distribution to theoretical Binomial.

flips = np.random.choice([0,1], size=1000, p=[0.5,0.5])
heads_count = np.sum(flips)

print("Heads in 1000 flips:", heads_count)

plt.hist(flips, bins=2, density=True, rwidth=0.8)
plt.xticks([0,1], ["Tails","Heads"])
plt.title("Simulated Coin Flips")
plt.show()
```

# Key Takeaways

Random variables let us connect outcomes to numbers.

Distributions describe how likely values are.

Binomial (discrete) and Normal (continuous) are the most important for ML.

Normal distribution underpins many algorithms (e.g., linear regression errors are assumed normal).
