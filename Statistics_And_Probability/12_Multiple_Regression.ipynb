### 1. Essence + Use Case + Example
#### Multiple Regression

*   **Essence:** Extend linear regression to multiple predictors.

*   **Use Case:** Predict outcomes with multiple factors; e.g., predicting exam scores from hours studied, sleep hours, and past performance.

*   **Example:**

    $Exam\_Score = m_1 * StudyHours + m_2 * SleepHours + m_3 * PastScore + b$

*   **Key Terms**

    *   **Coefficients ($m_1, m_2...$):** How much each X affects Y.
    *   **Intercept ($b$):** Y when all Xs = 0.
    *   **RÂ²:** How much variation in Y is explained by all Xs together.
    *   **Feature Analysis:** Determines which variables are most important for predicting Y.

---

### 2. Theory

*   **Equation:**

    $Y = m_1 X_1 + m_2 X_2 + \dots + m_n X_n + b$

*   **Goal:** Minimize sum of squared errors.

*   **High RÂ²:** Model explains more variance.

*   **Low RÂ²:** Model does not explain variance well.

### 3. Code Demonstration

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Example dataset
data = pd.DataFrame({
    'StudyHours': [1, 2, 3, 4, 5],
    'SleepHours': [7, 6, 8, 5, 7],
    'PastScore': [65, 70, 75, 80, 85],
    'ExamScore': [60, 70, 75, 80, 85]
})

# Features (X) and target (Y)
X = data[['StudyHours', 'SleepHours', 'PastScore']]
y = data['ExamScore']

# Fit multiple linear regression
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("RÂ² score:", model.score(X, y))

# Optional: visualize predicted vs actual
import matplotlib.pyplot as plt
plt.scatter(y, y_pred)
plt.xlabel("Actual Exam Score")
plt.ylabel("Predicted Exam Score")
plt.title("Multiple Regression Predictions")
plt.plot([min(y), max(y)], [min(y), max(y)], color='red')  # perfect fit line
plt.show()
```

## Checking Feature Importance by RÂ² Change

One way to understand the importance of each feature in a multiple linear regression model is to observe how the model's performance changes when a specific feature is removed. By comparing the RÂ² score of the full model (with all features) to the RÂ² score of models trained with one feature removed, we can infer the importance of the removed feature. A larger drop in RÂ² indicates that the removed feature was more important for the model's predictive power.

**Reasoning**:
The previous attempts to create a markdown cell failed. This time, add a new code cell to implement the feature importance check by iterating through each feature, training a model without it, and comparing the RÂ² with the full model's RÂ².



```python
# Calculate RÂ² of the full model (using student performance data)
X_student = student_data[['StudyHours', 'SleepHours', 'Attendance']]
y_student = student_data['Performance']
full_model = LinearRegression()
full_model.fit(X_student, y_student)
full_r2 = full_model.score(X_student, y_student)

print(f"RÂ² of the full model: {full_r2:.4f}\n")
print("Feature Importance (measured by RÂ² drop):")

# Iterate through each feature
for feature in X_student.columns:
    # Create a new feature set by dropping the current feature
    X_reduced = X_student.drop(columns=[feature])

    # Train a new LinearRegression model with the reduced feature set
    reduced_model = LinearRegression()
    reduced_model.fit(X_reduced, y_student)

    # Calculate the RÂ² of the new model
    reduced_r2 = reduced_model.score(X_reduced, y_student)

    # Calculate the drop in RÂ²
    r2_drop = full_r2 - reduced_r2

    # Print the results
    print(f"- Removing '{feature}': RÂ² = {reduced_r2:.4f}, RÂ² Drop = {r2_drop:.4f}")
```
