### Finding the Area of a Circle Using Rings

Imagine we want the area of a circle with radius R.  
Instead of thinking of the whole circle at once, break it into many thin **rings** (like onion layers).  

- Each ring has a **small thickness dr**.  
- The circumference of a ring at radius r is **2πr**.  
- So the area of one thin ring is approximately **2πr · dr**.  

Now, if we add up (integrate) all these rings from r = 0 to r = R, we get the total area:

A = ∫[0→R] 2πr dr = πR²  

This is the same formula we know for the area of a circle, but now we’ve *derived it* using calculus.


```python
import numpy as np
import matplotlib.pyplot as plt

# Radius of circle
R = 5

# Numerical approximation of area using many thin rings
dr = 0.001
'''The numpy.arange() function generates an array of evenly
   spaced values within a specified interval.'''
radii = np.arange(0, R, dr)
ring_areas = 2 * np.pi * radii * dr
approx_area = np.sum(ring_areas)

# Exact area
exact_area = np.pi * R**2

print("Approximated Area (rings method):", approx_area)
print("Exact Area (πR²):", exact_area)

# Visualization: Rings
fig, ax = plt.subplots()
circle = plt.Circle((0, 0), R, color='skyblue', alpha=0.3)
ax.add_patch(circle)

# Draw some sample rings
for r in np.linspace(0, R, 10):
    ring = plt.Circle((0, 0), r, color='blue', fill=False, alpha=0.5)
    ax.add_patch(ring)

ax.set_aspect('equal')
ax.set_xlim(-R-1, R+1)
ax.set_ylim(-R-1, R+1)
plt.title("Circle Approximated by Rings")
plt.show()
```

### Essence of Calculus

At its heart, calculus is about breaking down problems into **tiny pieces** and then adding them back up.

1. **Differentiation (Rates of Change)**  
   - Ask: "How fast is something changing at this very instant?"  
   - Example: Speed is the rate of change of distance.  
   - Mathematically: f'(x) = lim(h→0) [f(x+h) - f(x)] / h  

2. **Integration (Summing Tiny Pieces)**  
   - Ask: "What happens if we add up infinitely many tiny contributions?"  
   - Example: Area under a curve, total mass, accumulated distance.  
   - Mathematically: ∫ f(x) dx ≈ sum of f(x)·Δx as Δx → 0  

3. **Unifying Idea**  
   - Differentiation splits big changes into small ones.  
   - Integration combines small changes into big results.  
   - These two are **inverse operations** (Fundamental Theorem of Calculus).  


```python
import numpy as np

# defining a function

def f(x):
  return x**2

# Derrivative (slope at a point) using small difference
def derrivative(x, h=1e-5): # 1e-5 = 0.00001
  return (f(x+h) - f(x)) / h

# Integral (area under curve) using Riemann sum
def integral(a,b,n=1000):
  x = np.linspace(a,b,n)
  dx = x[1] - x[0]
  return np.sum(f(x) * dx)

# results
print("Derrivative at x=1:", derrivative(1))
print("Integral from 0 to 2:", integral(0,2))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function
f = lambda x: x**2   # simple parabola
df_approx = lambda x, h=1e-5: (f(x+h) - f(x)) / h  # numerical derivative

# Points to test
x = np.linspace(0, 2, 100)
y = f(x)

# Numerical derivative at a point
point = 1.0
slope = df_approx(point)

# Integral approximation (Riemann sum)
dx = 0.01
area = np.sum(f(x) * dx)

print("Derivative of f(x)=x^2 at x=1 ≈", slope)
print("Approximate integral of f(x)=x^2 from 0 to 2 ≈", area)

# Visualization
plt.plot(x, y, label="f(x)=x²")
plt.plot(point, f(point), 'ro', label="Point of derivative")
plt.plot(x, slope*(x-point)+f(point), 'g--', label="Tangent line at x=1")
plt.fill_between(x, y, where=(x>=0) & (x<=2), alpha=0.3, label="Approx Area (Integral)")
plt.legend()
plt.title("Essence of Calculus: Derivative = slope, Integral = area")
plt.show()
```
